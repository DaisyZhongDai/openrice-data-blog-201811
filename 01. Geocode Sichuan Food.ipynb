```python
from geopy.geocoders import GoogleV3
geolocator = GoogleV3(api_key='')

def geo_encode(x):
    #geolocator.geocode
    #print(x['location'])
    #print(x['point'])
    try:
        if x['point'] is None:
            x['loc'] = geolocator.geocode(x['location'])
#             print(x['loc'])
            if x['loc']:
                x['point'] = tuple(x['loc'].point)
                print(x['location'], x['point'])            
#                 pass
    except Exception as e:
        print(e)
    return x
# from geopy.extra.rate_limiter import RateLimiter
# geocode = RateLimiter(geolocator.geocode, min_delay_seconds=0.2)
df_sichuan_geo = df_sichuan.apply(geo_encode, axis=1)
```

```python
len(df_sichuan_geo)
```

```python
df_sichuan_geo['latitude'] = df_sichuan_geo['point'].apply(lambda x: x[0] if x else None)
df_sichuan_geo['longitude'] = df_sichuan_geo['point'].apply(lambda x: x[1] if x else None)
```

```python
df_sichuan_geo.to_csv('sichuan.csv')
```
